## Data Visualization With Seaborn

```python
import seaborn as sns
```

```python
# Basic plotting with seaborn
tips= sns.load_dataset('tips')
tips
```

```python
import matplotlib.pyplot as plt
```

```python
sns.scatterplot(data=tips)
```

Scatter plot

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Scatter Plot of Total Bill vs Tips')
plt.show()
```

Line Plot

```python

sns.lineplot(x='size',y='tip', data=tips,label='Tips',marker='X', markersize=6)

sns.lineplot(x='size', y='total_bill', data=tips, label='Total Bill', marker='^', markersize=6)

plt.title('tips and total bill with respect to size')
plt.legend(title='Metrics')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(45,30))

plt.subplot(10,8,1)
sns.lineplot(x='size',y='tip', data=tips,)
plt.title('Tips vs group size')
plt.grid()

plt.subplot(10,8,2)
sns.lineplot(x='size', y='total_bill', data=tips)
plt.title('Tips vs total bill')
plt.grid()

plt.show()
```

#### Plotting on real data

```python
sales_df= pd.read_csv('sales_data.csv')
sales_df.dtypes
```

Ploting total sales by product

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Product Category', y='Total Revenue', data= sales_df, estimator=sum)
plt.title('Revenue per product category')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(y='Total Revenue', x='Region',data=sales_df, estimator=sum)
plt.xlabel('Region')
plt.ylabel('Revenue')
plt.title('Revenue per Region')
plt.show()
```
