#### Data aggregation and Grouping

```python
df.groupby("Product").agg({
    "Date": "first",          # or 'min', 'max'
    "Category": "first",      # or 'unique'
    "Value": "mean",          # numeric → mean
    "Sales": "sum",           # numeric → sum
    "Region": "first",        # or 'unique'
    "Sales_fillNA": "mean",
    "Value_new": "mean"
})

```

```python
df.groupby("Product").agg(list)
```

```python
df_product_group=df.groupby("Product")[["Value",'Sales']].agg(["sum", "mean", "max"])
```

#### Merging and joining data frames

```python
df1= pd.DataFrame({'key':['A','B','C'], 'Value':[1,2,3]})
df2= pd.DataFrame({'key':['A','B','D'], 'Value':[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
df_m=pd.merge(df1,df2, on='key',how='inner')
df_m
```

```python
df_fm=pd.merge(df1,df2, on='key', how='outer')
df_fm
```

```python
# df['Value_new']=df['Value'].fillna(df['Value'].mean()).astype(int)
# df.head(5)
df_fm['Value_x']= df_fm['Value_x'].fillna(df_fm['Value_x'].mean())
df_fm
```

```python
df_fm['Value_y']=df_fm['Value_y'].fillna(df_fm['Value_y'].mean())
df_fm
```

```python
df_lfm=pd.merge(df1,df2,on='key', how='left')
df_lfm
```

```python
df_rfm=pd.merge(df1,df2,on='key', how='right')
df_rfm
```
