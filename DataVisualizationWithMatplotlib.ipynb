## Data Visualization with MathPlotLib

```python
import matplotlib.pyplot as plt
```

Line Chart

```python
x=[1,23,32,14,43,64,57,58,91,10]
y=[1,10,15,20,25,30,35,40,45,50]
print('x ',len(x))
print('y ',len(y))
```

```python
#create a line plot
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Testing testing 123...')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
# Creating a custom line plot
plt.plot(x,y, color='red',linestyle='--', marker='o',mfc='blue', markersize='10')
plt.grid(True)
plt.show()
```

### Multi plot

```python
# Sample data
x=[1,2,3,4,5,6]
y1=[11,21,13,24,5,5]
y2=[10,12,23,14,25,10]

plt.figure(figsize=(9,7))

# PLot 1
plt.subplot(2,2,1)
plt.plot(x,y1, color='green', marker='o')
plt.title('Plot 1')
plt.grid(True)

# PLot 2
plt.subplot(2,2,2)
plt.plot(x,y2, color='red', marker='o')
plt.title('Plot 2')
plt.grid(True)

# Plot 3
plt.subplot(2,2,3)
plt.plot(x,y1, color='blue', marker='o')
plt.title('Plot 3')
plt.grid(True)

# Plot 4
plt.subplot(2,2,4)
# x2=([(((i/2)*4) if i%2==0 else i *2 ) for i in x ])
# print(x2)
plt.plot(([(((i/2)*4) if i%2==0 else i *2 ) for i in x ]),y1, color='orange', marker='o')
plt.title('Plot 4')
plt.grid(True)

```
